```python
# Menampilkan Data urutan Awal
# Membaca dataset
airbnb_data = pd.read_csv("Airbnb Listings Bangkok.csv")

# Menampilkan lima baris pertama dari dataset untuk melihat strukturnya
airbnb_data.head()
```

Mengidentifikasi Kolom dengan Nilai yang Hilang:

1.   Menghitung jumlah nilai yang hilang di setiap kolom.
2.   Menangani Nilai yang Hilang:


```python
# Menghitung jumlah nilai yang hilang di setiap kolom
missing_values = airbnb_data.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Mengisi nilai yang hilang pada kolom 'reviews_per_month' dengan 0
airbnb_data['reviews_per_month'].fillna(0, inplace=True)

# Mengganti nilai yang hilang pada kolom 'last_review' dengan NaT (Not a Time)
airbnb_data['last_review'] = pd.to_datetime(airbnb_data['last_review'])
airbnb_data['last_review'].fillna(pd.NaT, inplace=True)

# Mengisi nilai yang hilang pada kolom 'host_name' dengan 'Unknown'
airbnb_data['host_name'].fillna('Unknown', inplace=True)

# Mengisi nilai yang hilang pada kolom 'name' dengan 'No Name'
airbnb_data['name'].fillna('No Name', inplace=True)
```

Langkah selanjutnya adalah mengonversi kolom `last_review` ke format datetime untuk analisis lebih lanjut, serta mengidentifikasi dan menangani outliers, terutama pada kolom `price` dan `number_of_reviews`, untuk memastikan bahwa data yang ekstrem tidak mempengaruhi analisis. ### 4. Mengidentifikasi dan Menangani Outliers


1.   Mengidentifikasi Outliers
2.   Menangani Outliers



```python
# Mengidentifikasi outliers menggunakan boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=airbnb_data['price'])
plt.title('Boxplot of Price')
plt.show()

# Menghapus outliers pada kolom 'price' (mengasumsikan outliers berada di atas persentil 99)
upper_limit = airbnb_data['price'].quantile(0.99)
airbnb_data = airbnb_data[airbnb_data['price'] <= upper_limit]
```

Dari boxplot yang dihasilkan, terlihat bahwa ada sejumlah outliers pada kolom `price` yang perlu ditangani. Langkah selanjutnya adalah menghapus outliers tersebut dan memeriksa kembali distribusi harga setelah outliers dihapus. Setelah itu, kita akan melanjutkan dengan analisis data eksploratif (EDA) untuk memahami lebih lanjut tentang dataset ini.

```python
# Menghapus outliers pada kolom 'price' (mengasumsikan outliers berada di atas persentil 99)
upper_limit = airbnb_data['price'].quantile(0.99)
airbnb_data = airbnb_data[airbnb_data['price'] <= upper_limit]

# Menampilkan boxplot setelah menghapus outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=airbnb_data['price'])
plt.title('Boxplot of Price (After Removing Outliers)')
plt.show()

# Menampilkan histogram untuk memeriksa distribusi harga setelah menghapus outliers
plt.figure(figsize=(10, 6))
sns.histplot(airbnb_data['price'], kde=True)
plt.title('Histogram of Price (After Removing Outliers)')
plt.show()
```

### 6. Menangani Nama yang Aneh dengan Mengganti dengan "Unique Name"

Dalam dataset Airbnb Listings Bangkok, kita menemukan bahwa beberapa entri di kolom `name` dan `host_name` mengandung karakter aneh atau non-ASCII. Karakter-karakter ini dapat disebabkan oleh encoding yang salah atau input data yang tidak konsisten. Untuk memastikan analisis data yang akurat dan konsisten, kita akan mengganti semua entri yang mengandung karakter aneh dengan teks "Unique Name". Ini akan membantu kita menghindari potensi masalah selama analisis lebih lanjut.

```python
import re

# Fungsi untuk memeriksa dan mengganti teks yang mengandung karakter non-ASCII dengan "Unique Name"
def replace_unique_name(text):
    if isinstance(text, str):
        if re.search(r'[^\x00-\x7F]+', text):
            return "Unique Name"
    return text

# Menerapkan fungsi replace_unique_name ke kolom 'name' dan 'host_name' pada dataset salinan
airbnb_data['name'] = airbnb_data['name'].apply(replace_unique_name)
airbnb_data['host_name'] = airbnb_data['host_name'].apply(replace_unique_name)

# Menampilkan beberapa baris pertama setelah pembersihan
airbnb_data[['name', 'host_name']].head(5)
```

# B. **Exploratory Data Analysis (EDA)**

## 1. Mengidentifikasi Lokasi Strategis

### 1.1 Distribusi Listing Berdasarkan Lokasi
Memeriksa distribusi listing berdasarkan lingkungan (`neighbourhood`) untuk mengidentifikasi lokasi strategis.


```python
# 1.1 Distribusi Listing Berdasarkan Lokasi
plt.figure(figsize=(12, 8))
airbnb_data['neighbourhood'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Neighbourhoods by Number of Listings')
plt.xlabel('Neighbourhood')
plt.ylabel('Number of Listings')
plt.show()
```

### 1.2 Distribusi Harga Berdasarkan Lokasi
Memeriksa distribusi harga listing di berbagai lingkungan untuk menemukan lokasi strategis dengan harga yang menguntungkan.

```python
# 1.2 Distribusi Harga Berdasarkan Lokasi
plt.figure(figsize=(14, 8))
sns.boxplot(x='neighbourhood', y='price', data=airbnb_data)
plt.xticks(rotation=90)
plt.title('Price Distribution by Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Price')
plt.show()
```

### 1.3 Peta Distribusi Listing Berdasarkan Lokasi
Membuat peta sebaran listing berdasarkan koordinat (latitude dan longitude) untuk visualisasi geografis. Ini membantu mengidentifikasi konsentrasi listing di berbagai bagian kota.

```python
import geopandas as gpd
from shapely.geometry import Point

# Membuat GeoDataFrame dari data listing
geometry = [Point(xy) for xy in zip(airbnb_data['longitude'], airbnb_data['latitude'])]
gdf = gpd.GeoDataFrame(airbnb_data, geometry=geometry)

# Membuat peta sebaran listing
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world[world.name == 'Thailand'].plot(color='white', edgecolor='black')

# Plot GeoDataFrame
gdf.plot(ax=ax, markersize=5, color='blue', alpha=0.5)
plt.title('Geographical Distribution of Listings in Bangkok')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### 1.4 Menghitung Rata-rata Harga dan Jumlah Ulasan per Lingkungan
Menghitung rata-rata harga dan jumlah ulasan per lingkungan (neighbourhood). Analisis ini membantu kita memahami lingkungan mana yang memiliki harga sewa rata-rata tertinggi dan lingkungan mana yang paling banyak diulas.

```python
# Menghitung rata-rata harga dan jumlah ulasan per lingkungan
neighbourhood_analysis = airbnb_data.groupby('neighbourhood').agg({
    'price': 'mean',
    'number_of_reviews': 'sum'
}).reset_index()

# Mengurutkan lingkungan berdasarkan harga rata-rata
neighbourhood_analysis = neighbourhood_analysis.sort_values(by='price', ascending=False)

# Menampilkan 10 lingkungan teratas berdasarkan harga rata-rata
print("Top 10 Neighbourhoods by Average Price:")
print(neighbourhood_analysis.head(10))

# Plotting 10 lingkungan teratas berdasarkan harga rata-rata
plt.figure(figsize=(12, 8))
sns.barplot(x='price', y='neighbourhood', data=neighbourhood_analysis.head(10), palette='viridis')
plt.title('Top 10 Neighbourhoods by Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.show()
```

### 1.5 Menghitung Jumlah Ulasan per Lingkungan


```python
# Mengurutkan lingkungan berdasarkan jumlah ulasan
neighbourhood_analysis = neighbourhood_analysis.sort_values(by='number_of_reviews', ascending=False)

# Menampilkan 10 lingkungan teratas berdasarkan jumlah ulasan
print("Top 10 Neighbourhoods by Number of Reviews:")
print(neighbourhood_analysis.head(10))

# Plotting 10 lingkungan teratas berdasarkan jumlah ulasan
plt.figure(figsize=(12, 8))
sns.barplot(x='number_of_reviews', y='neighbourhood', data=neighbourhood_analysis.head(10), palette='magma')
plt.title('Top 10 Neighbourhoods by Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Neighbourhood')
plt.show()
```

### 1.6 Menghitung Rasio Harga terhadap Ulasan


```python
# Menghitung rasio harga terhadap ulasan
neighbourhood_analysis['price_to_review_ratio'] = neighbourhood_analysis['price'] / (neighbourhood_analysis['number_of_reviews'] + 1)

# Mengurutkan lingkungan berdasarkan rasio harga terhadap ulasan
neighbourhood_analysis = neighbourhood_analysis.sort_values(by='price_to_review_ratio', ascending=False)

# Menampilkan 10 lingkungan teratas berdasarkan rasio harga terhadap ulasan
print("Top 10 Neighbourhoods by Price to Review Ratio:")
print(neighbourhood_analysis.head(10))

# Plotting 10 lingkungan teratas berdasarkan rasio harga terhadap ulasan
plt.figure(figsize=(12, 8))
sns.barplot(x='price_to_review_ratio', y='neighbourhood', data=neighbourhood_analysis.head(10), palette='coolwarm')
plt.title('Top 10 Neighbourhoods by Price to Review Ratio')
plt.xlabel('Price to Review Ratio')
plt.ylabel('Neighbourhood')
plt.show()
```

```python
# Calculating the average price per neighbourhood
avg_price_neighbourhood = airbnb_data.groupby('neighbourhood')['price'].mean().reset_index()

# Sorting the neighbourhoods by average price
avg_price_neighbourhood = avg_price_neighbourhood.sort_values(by='price', ascending=False)

# Plotting the average price per neighbourhood
plt.figure(figsize=(14, 8))
sns.barplot(x='price', y='neighbourhood', data=avg_price_neighbourhood.head(10), palette='viridis')
plt.title('Top 10 Neighbourhoods by Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.show()
```

```python
# Plotting the correlation between price and number of reviews
plt.figure(figsize=(14, 8))
sns.scatterplot(x='price', y='number_of_reviews', data=airbnb_data, hue='room_type', palette='Set2')
plt.title('Correlation Between Price and Number of Reviews')
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.legend(title='Room Type')
plt.show()
```

```python
# Ensuring 'last_review' column is datetime
airbnb_data['last_review'] = pd.to_datetime(airbnb_data['last_review'], errors='coerce')

# Extracting month and year from the 'last_review' column
airbnb_data['year'] = airbnb_data['last_review'].dt.year
airbnb_data['month'] = airbnb_data['last_review'].dt.month

# Calculating the average price per month
avg_price_month = airbnb_data.groupby('month')['price'].mean().reset_index()

# Plotting the average price per month
plt.figure(figsize=(14, 8))
sns.lineplot(x='month', y='price', data=avg_price_month, marker='o', palette='coolwarm')
plt.title('Seasonal Trend in Average Price')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(range(1, 13))
plt.show()
```

## 5. Analisis Tren dan Musiman

### 5.1 Tren Rata-rata Harga Bulanan
Kode ini akan memvisualisasikan tren rata-rata harga bulanan untuk listing Airbnb di Bangkok. Analisis ini membantu untuk memahami bagaimana harga sewa bervariasi seiring waktu dan apakah ada pola musiman dalam penetapan harga.


```python
# Grouping by year and month to calculate the average price per month
monthly_avg_price = airbnb_data.groupby(['year', 'month'])['price'].mean().reset_index()

# Plotting the trend of monthly average prices
plt.figure(figsize=(14, 8))
sns.lineplot(x='month', y='price', hue='year', data=monthly_avg_price, marker='o', palette='tab10')
plt.title('Monthly Average Price Trends Over Years')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.legend(title='Year')
plt.show()
```

### 5.2 Tren Rata-rata Ulasan per Bulan
Kode ini akan memvisualisasikan tren rata-rata ulasan per bulan untuk setiap tipe kamar yang tersedia di Airbnb Bangkok. Analisis ini membantu untuk melihat apakah ada perbedaan signifikan dalam tren ulasan bulanan berdasarkan tipe kamar yang disewa.

```python
# Calculating the average reviews per month per room type by year and month
avg_reviews_per_month_room_type = airbnb_data.groupby(['year', 'month', 'room_type'])['reviews_per_month'].mean().reset_index()

# Plotting the trend of average reviews per month per room type
plt.figure(figsize=(14, 8))
sns.lineplot(x='month', y='reviews_per_month', hue='room_type', data=avg_reviews_per_month_room_type, marker='o', palette='Set2')
plt.title('Monthly Average Reviews per Room Type')
plt.xlabel('Month')
plt.ylabel('Average Reviews per Month')
plt.legend(title='Room Type')
plt.show()
```
